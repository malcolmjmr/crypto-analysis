```python
class TrelloApi(object):

    def __init__(self, url=API_URL, token=API_TOKEN, secret=API_SECRET):
        self.api_url = url
        self.credentials = {'key': token, 'token': secret}
    
    def get_boards(self, params={'filter':'all'}):
        params.update(self.credentials)
        res = requests.get(f'{self.api_url}members/me/boards', params=params).json()
        self.boards = {b['name']: b for b in res}
        return self.boards 
    
    def select_board(self, board_name):
        if not hasattr(self, 'boards'):
            self.get_boards()   
        self.current_board = self.boards[board_name]
        
    def get_lists(self):
        assert hasattr(self, 'current_board')
        board_id = self.current_board['id']
        res = requests.get(f'{self.api_url}boards/{board_id}/lists', params=self.credentials).json()
        self.current_board['lists'] = json_normalize(res)
        return self.current_board['lists']
        
    def get_cards(self, board_id=self.current_board['id']):
        assert hasattr(self, 'current_board')
        res = requests.get(f'{self.api_url}boards/{board_id}/cards', params=self.credentials).json()
        self.current_board['cards'] = json_normalize(res)
        return self.current_board['cards']
    
    def get_actions(obj_id):
        
        


    
```
